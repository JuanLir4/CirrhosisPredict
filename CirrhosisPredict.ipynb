```python
from sklearn.metrics import confusion_matrix
```

```python
#temos 3 classificações de edema:
#N (no edema and no diuretic therapy for edema), 
#S (edema present without diuretics, or edema resolved by diuretics), 
#Y (edema despite diuretic therapy)

#vamos criar uma matriz de confusão e ver como esses dados se comportam

# Mapeamento dos status para categorias numéricas para uso na matriz de confusão
status_mapping1 = {'C': 0, 'CL': 1, 'D': 2}
status_mapping2 = {'N': 0, 'S': 1, 'Y': 2}

# Defina as variáveis alvo e previstas
y_true = data_cr5['Edema'].map(status_mapping2)
y_pred = data_cr5['Status'].map(status_mapping1)

# Crie a matriz de confusão
conf_matrix = confusion_matrix(y_true, y_pred)

# Converta a matriz de confusão em um DataFrame pandas para uma visualização mais clara
confusion_df = pd.DataFrame(conf_matrix, index=['N', 'S', 'Y'], columns=['C', 'CL', 'D'])

print("Matriz de Confusão:")
display(confusion_df)
```

```python
#OBSERVAMOS QUE PACIENTES SEM EDEMA SAO BEM MAIS COMUNS,
#E QUE AO EDEMA PERSISTIR A TERAPIA COM DIURETICOS, A CHANCE DE MORTE É MUITO GRANDE
```

```python
data_cr5['Stage'].unique()
#representam em que grau da cirrose o paciente esta
```

```python
#Matriz de confusão:
status_mapping3 = {'C': 0, 'CL': 1, 'D': 2}

# Defina o mapeamento de status
status_mapping4 = {1: 0, 2 : 1, 3 : 2, 4 : 3}


y_true1 = data_cr5['Stage'].map(status_mapping4)  
y_pred1 = data_cr5['Status'].map(status_mapping3)

# Crie a matriz de confusão
conf_matrix = confusion_matrix(y_true1, y_pred1)

#removendo a coluna a mais que a função criou
conf_matrix1 = conf_matrix[:, :-1]



# Converta a matriz de confusão em um DataFrame pandas para uma visualização mais clara
confusion_df = pd.DataFrame(conf_matrix1, index=[1 , 2 , 3 , 4], columns=['C', 'CL', 'D'])

print("Matriz de Confusão:")
display(confusion_df)
```

```python
#pode ser observar muitas coisas atraves dessa matriz, os casos de morte aumentam muito a medida que o estagio da doença avança
#pretendo fazer uma analise em relação as duas matrizes que fiz, incluindo graficos.
```

```python
X_train = data_cr5[['Drug', 'Age', 'Sex', 'Ascites','Hepatomegaly', 'Spiders', 'Edema', 'Bilirubin', 'Cholesterol','Albumin', 'Copper', 'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets','Prothrombin', 'Stage']]
Y_train = data_cr5['Status']
```

```python
X_train.reset_index(drop=True, inplace=True)
Y_train.reset_index(drop=True, inplace=True)
```

```python
#colocar certas variaveis em onehot
#são elas: 'Drug', 'Sex', 'Ascites','Hepatomegaly', 'Spiders', 'Edema' e a propria saida(y)

from sklearn.preprocessing import LabelEncoder, OneHotEncoder 

labelencoder = LabelEncoder()

#colocando em numero:


Y_train = labelencoder.fit_transform(Y_train)
#voltando para datset:
Y_train = pd.DataFrame(Y_train)
```

```python
Y_train.head()
```

```python
# Listar as colunas que você deseja codificar
#Muito cuidado nessa parte, nao sei se ele embaralha os indices
colunas = ['Drug', 'Sex', 'Ascites', 'Hepatomegaly', 'Spiders', 'Edema']


# Iterar sobre cada coluna, transformar em um array individual e aplicar o LabelEncoder
for coluna in colunas:
    array_individual = X_train.loc[: , coluna].values  # Converter a coluna em um array individual
    X_train.loc[:, coluna] = labelencoder.fit_transform(array_individual)  # Aplicar o LabelEncoder e atribuir de volta ao DataFrame
```

```python
#agora, precisamos convertar em vetores:
from keras.utils import to_categorical
#faremos isso utlizando o "to_categorical" que aceita apenas matrizes com uma dimensão, entao usaremos a msm ideia de cima
```

```python
Y_train = to_categorical(Y_train)
Y_train = pd.DataFrame(Y_train)
Y_train.head()
```

```python
from keras.utils import to_categorical

array_individual = X_train.loc[:, 'Edema']
array_individual_categorical = to_categorical(array_individual)
array_individual_categorical = pd.DataFrame(array_individual_categorical, columns=['Edema0', 'Edema1', 'Edema2'])

array_individual_categorical.head()

#como 'array_individual_categorica' tem 3 colunas, precisamos concatenar com o dataframe anterior e depois excluir a antiga coluna
```

```python
X_train.shape
```

```python
X_trainEdema = pd.concat([X_train, array_individual_categorical], axis=1)

X_trainEdema.head()
```

```python
status_mapping4 = {1: 0, 2 : 1, 3 : 2, 4 : 3}
array_stage = X_train.loc[:, 'Stage'].map(status_mapping4)


array_stage_individual = to_categorical(array_stage)

array_stage_categorical = pd.DataFrame(array_stage_individual , columns=['Stage1', 'Stage2', 'Stage3', 'Stage4'])

array_stage_categorical.head()
```

```python
array_stage_categorical.shape
```

```python
X_trainTotal = pd.concat([X_trainEdema, array_stage_categorical], axis=1)
X_trainTotal.head()
   
```

```python
del X_trainTotal['Stage']
X_trainTotal.head(20)
```

```python
#começar a montar a rede neural:
X_trainTotal.tail()
```

```python
from sklearn.model_selection import train_test_split
X_treino, X_teste, Y_Treino, Y_teste = train_test_split(X_trainTotal, Y_train, random_state = 17, test_size = 0.2)
```

```python
from keras.layers import Dense,Dropout
from keras.models import Sequential
```

```python
#pretendo mexer aqui e ajustar alguns parametros
#apesar de q o conjunto de dados é muito pequeno
classificador = Sequential()

classificador.add(Dense(units = 24, activation = 'relu'))
classificador.add(Dense(units = 12, activation = 'relu'))
classificador.add(Dense(units = 6, activation = 'relu'))
classificador.add(Dense(units = 3, activation = 'softmax'))

classificador.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
                      metrics = ['categorical_accuracy'])

classificador.fit(X_treino, Y_Treino, batch_size = 100,
                  epochs = 1000)
```

```python
import numpy as np 

resultado = classificador.evaluate(X_treino, Y_Treino)
previsoes = classificador.predict(X_teste)
previsoes = (previsoes > 0.5)

Y_teste_numpy =Y_teste.values

classe_teste2 = [np.argmax(t) for t in Y_teste_numpy]
previsoes2 = [np.argmax(t) for t in previsoes]


matriz1 = confusion_matrix(previsoes2, classe_teste2)

confusion_df = pd.DataFrame(matriz1, index=['P.C', 'P.CL', 'P.D'], columns=['C', 'CL', 'D'])
display(confusion_df)
```
